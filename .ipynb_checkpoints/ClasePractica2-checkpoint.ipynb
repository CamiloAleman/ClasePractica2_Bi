```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
#Url del dataset

url = 'https://raw.githubusercontent.com/CamiloAleman/ClasePractica2_Bi/refs/heads/main/data/UltimateClassicRock.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

Glosario de variables

Track: The title of the track.
Artist: The name of the artist.
Album: The album in which the track appears.
Year: The release year of the album.
Duration: The length of the track in seconds.
Time_Signature: The time signature of the track (e.g., 4/4, 3/4).
Danceability: A measure of how suitable the track is for dancing, ranging from 0 to 1.
Energy: The intensity and activity level of the track, ranging from 0 to 1.
Key: The musical key in which the track is composed (e.g., 0=C).
Loudness: The overall loudness of the track in decibels (dB).
Mode: The modality of the track (e.g., major or minor).
Speechiness: The presence of spoken words in the track, ranging from 0 to 1.
Acousticness: The amount of acoustic sound in the track, ranging from 0 to 1.
Instrumentalness: The likelihood that the track is instrumental, ranging from 0 to 1.
Liveness: The presence of an audience or live performance elements, ranging from 0 to 1.
Valence: A measure of the track's positivity or happiness, ranging from 0 to 1.
Tempo: The speed of the track in beats per minute (BPM).
Popularity: The track's popularity score, usually based on streaming data and listener engagement.

Tratamiento de Outliers

```python
df = df[(np.abs(stats.zscore(df['Danceability'])) < 3)]
```

Transformación de Variables

```python
scaler = StandardScaler()

df['Danceability_scaled'] = scaler.fit_transform(df[['Danceability']])
```

```python
min_max_scaler = MinMaxScaler()

df['Danceability_normalized'] = min_max_scaler.fit_transform(df[['Danceability']])

print(df[['Danceability', 'Danceability_scaled', 'Danceability_normalized']].head())
```

Ingeniería de Características

```python
df['GrLivArea_per_room'] = df['GrLivArea'] / df['TotRmsAbvGrd']
df['Price_per_sqft'] = df['SalePrice'] / df['GrLivArea']

print(df[['GrLivArea_per_room', 'Price_per_sqft']].head())
```
